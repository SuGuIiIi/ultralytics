```python
# prompt: 读取europe_emissions.csv中的各个国家，并检测各个国家的空值，选取空值最少的25个国家，生成新的表格

import pandas as pd

# 读取欧洲排放数据
try:
    europe_df = pd.read_csv('europe_emissions.csv')
except FileNotFoundError:
    print("Error: 'europe_emissions.csv' not found. Please run the previous code block first.")
    exit()

# 计算每个国家的数据缺失值数量
missing_values_by_country = europe_df.isnull().sum(axis=1)

# 将缺失值数量添加到数据框中
europe_df['Missing_Values'] = missing_values_by_country

# 按缺失值数量排序，并选择前25个国家
sorted_europe_df = europe_df.sort_values('Missing_Values')
top_25_countries = sorted_europe_df.head(25)

# 删除 'Missing_Values' 列（可选）
top_25_countries = top_25_countries.drop(columns=['Missing_Values'])

# 将结果保存到新的 CSV 文件
top_25_countries.to_csv('top_25_europe_emissions.csv', index=False)
print("Successfully created 'top_25_europe_emissions.csv'")
```

```python
# 读取数据
df = pd.read_csv('top_25_europe_emissions.csv')

# 获取排放量的列名
emission_cols = [col for col in df.columns if 'Metric tons of CO2e per capita' in col]

# 创建年份和统计量的DataFrame
years = [1990, 1995, 2000, 2005, 2010, 2013, 2018]
stats_data = []

for year in years:
    col_name = f'Metric tons of CO2e per capita ({year})'
    year_data = df[col_name].dropna()  # 去除NaN值

    stats_data.append({
        'Year': year,
        'Mean': round(year_data.mean(), 2),
        'Median': round(year_data.median(), 2),
        'Std Dev': round(year_data.std(), 2)
    })

# 创建结果DataFrame
results_df = pd.DataFrame(stats_data)

# 设置显示格式
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# 创建图表
plt.figure(figsize=(12, 6))

# 绘制均值和中位数的折线图
plt.plot(results_df['Year'], results_df['Mean'], marker='o', label='Mean', linewidth=2)
plt.plot(results_df['Year'], results_df['Median'], marker='s', label='Median', linewidth=2)

plt.title('Mean and Median CO2 Emissions per Capita Over Time')
plt.xlabel('Year')
plt.ylabel('Metric tons of CO2e per capita')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# 显示表格和图表
print("\nDescriptive Statistics for CO2 Emissions per Capita:")
print(results_df.to_string(index=False))
plt.show()

# 创建箱线图
plt.figure(figsize=(12, 6))
data_for_box = []
labels = []

for year in years:
    col_name = f'Metric tons of CO2e per capita ({year})'
    data_for_box.append(df[col_name].dropna())
    labels.append(str(year))

plt.boxplot(data_for_box, labels=labels)
plt.title('Distribution of CO2 Emissions per Capita')
plt.xlabel('Year')
plt.ylabel('Metric tons of CO2e per capita')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# 读取数据
df = pd.read_csv('top_25_europe_emissions.csv')

# 计算统计量
years = [1990, 1995, 2000, 2005, 2010, 2013, 2018]
stats_data = []

for year in years:
    col_name = f'Metric tons of CO2e per capita ({year})'
    year_data = df[col_name].dropna()

    stats_data.append({
        'Year': year,
        'Mean': round(year_data.mean(), 2),
        'Median': round(year_data.median(), 2),
        'Std Dev': round(year_data.std(), 2)
    })

results_df = pd.DataFrame(stats_data)

# 创建更专业的可视化
plt.figure(figsize=(12, 6))

# 绘制均值和中位数的线
plt.plot(results_df['Year'], results_df['Mean'], 'b-', marker='o', label='Mean', linewidth=2)
plt.plot(results_df['Year'], results_df['Median'], 'g--', marker='s', label='Median', linewidth=2)

# 添加标准差范围区域
plt.fill_between(results_df['Year'],
                 results_df['Mean'] - results_df['Std Dev'],
                 results_df['Mean'] + results_df['Std Dev'],
                 alpha=0.2, color='blue', label='±1 Std Dev')

# 设置图表样式
plt.title('CO₂ Emissions per Capita in European Countries (1990-2018)', fontsize=12, pad=15)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Metric tons of CO₂e per capita', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right')

# 设置y轴范围，确保能显示完整的标准差范围
y_min = min(results_df['Mean'] - results_df['Std Dev']) - 1
y_max = max(results_df['Mean'] + results_df['Std Dev']) + 1
plt.ylim(y_min, y_max)

# 添加数据标签
for i, row in results_df.iterrows():
    plt.annotate(f'{row["Mean"]:.1f}',
                (row['Year'], row['Mean']),
                textcoords="offset points",
                xytext=(0,10),
                ha='center')

plt.tight_layout()
plt.show()

# 打印统计表格
print("\nDescriptive Statistics for CO2 Emissions per Capita:")
print(results_df.to_string(index=False))
```

```python
# 读取数据
df = pd.read_csv('top_25_europe_emissions.csv')

# 计算统计量
years = [1990, 1995, 2000, 2005, 2010, 2013, 2018]
stats_data = []

for year in years:
    col_name = f'Metric tons of CO2e per capita ({year})'
    year_data = df[col_name].dropna()

    stats_data.append({
        'Year': year,
        'Mean': round(year_data.mean(), 2),
        'Median': round(year_data.median(), 2),
        'Std Dev': round(year_data.std(), 2)
    })

results_df = pd.DataFrame(stats_data)

# 创建两个子图
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12), height_ratios=[2, 1])

# 第一个图：趋势线
ax1.plot(results_df['Year'], results_df['Mean'], 'b-', marker='o', label='Mean', linewidth=2, markersize=8)
ax1.plot(results_df['Year'], results_df['Median'], 'g--', marker='s', label='Median', linewidth=2, markersize=8)

# 添加数据标签
for i, row in results_df.iterrows():
    ax1.annotate(f'Mean: {row["Mean"]:.1f}',
                (row['Year'], row['Mean']),
                textcoords="offset points",
                xytext=(0,15),
                ha='center',
                fontsize=10,
                bbox=dict(facecolor='none', edgecolor='none', alpha=0.7))

    ax1.annotate(f'Median: {row["Median"]:.1f}',
                (row['Year'], row['Median']),
                textcoords="offset points",
                xytext=(0,-25),
                ha='center',
                fontsize=10,
                bbox=dict(facecolor='none', edgecolor='none', alpha=0.7))

# 设置第一个图的样式
ax1.set_title('CO₂ Emissions Trends in European Countries (1990-2018)', fontsize=14, pad=20)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Metric tons of CO₂e per capita', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(loc='upper right', fontsize=12)

# 第二个图：标准差条形图
ax2.bar(results_df['Year'], results_df['Std Dev'], color='lightblue', alpha=0.9)
ax2.set_title('Standard Deviation of CO₂ Emissions', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Standard Deviation', fontsize=12)

# 确保y轴范围足够显示所有数值
max_std = results_df['Std Dev'].max()
ax2.set_ylim(0, max_std * 1.2)  # 设置y轴上限为最大值的1.2倍

# 添加标准差数值标签
for i, v in enumerate(results_df['Std Dev']):
    ax2.text(results_df['Year'][i], v, f'{v:.2f}',
             ha='center', va='bottom', fontsize=10)

ax2.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 打印统计表格
print("\nDescriptive Statistics for CO2 Emissions per Capita:")
print(results_df.to_string(index=False))
```
